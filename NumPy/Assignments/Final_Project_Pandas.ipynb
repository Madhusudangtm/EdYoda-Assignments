```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1: IMPORT THE DATASET USING PANDAS FROM ABOVE MENTIONED URL: 

Url = 'https://raw.githubusercontent.com/SR1608/Datasets/main/covid-data.csv'
df = pd.read_csv('https://raw.githubusercontent.com/SR1608/Datasets/main/covid-data.csv')
df.head()
```

```python
# CHECKING FOR NULL VALUES: 
df.isnull().sum()
```

```python
# 2 : A:  FIND THE NO OF ROWS AND COLUMNS IN THE DATASET: 
print('Total no of Row is: ',df.shape[0])
print('Total no of Columns is: ',df.shape[1])
```

```python
# 2: B: DATA TYPES OF COLUMNS: 
df.dtypes
```

```python
#2: c: INFO OF DATASET
df.info()
# 2: C: DESCRIBE OF THE DATA IN DATAFRAME: 
df.describe(include='all')
```

```python
# 2: C: DESCRIBE OF THE DATA IN DATAFRAME: 
df.describe()
```

```python
# 3: A: FIND COUNT OF UNIQUE VALUES IN LOCATION COLUMN.
print(f"Count of unique value in location column is {len(df['location'].value_counts())}")
```

```python
# 3: B: FIND WHICH CONTINENT HAS MAXIMUM FREQUENCY USING VALUES: 
print(f"\ncontinent having a maximum frequency is {df['continent'].value_counts().idxmax()}")
```

```python
# 3: C: FIND MAXIMUM AND MEAN VALUE IN TOTAL CASES:
x=df['total_cases'].max()
print("Maximum value: ",x)

y=df['total_cases'].mean()
print("Mean Value: ",y)
```

```python
# 3: D: FIND 25%, 50% AND 75% QUARTILE VALUE IN TOTAL DEATHS:

print(df['total_deaths'].quantile([0.25,0.5,.75]))
```

```python
#3: D: SECOND APPROACH 

print(f"25% of total_death_value: {df['total_deaths'].quantile(q=.25)}")
print(f"50% of total_death_value: {df['total_deaths'].quantile(q=.5)}")
print(f"75% of total_death_value: {df['total_deaths'].quantile(q=.75)}")
```

```python
# 3: E: FIND WHICH CONTINENT HAS MAXIMUM 'HUMAN_DEVELOPMENT_INDEX:
print(df.groupby('continent')['human_development_index'].max().idxmax())
print('-----------------------------------------------------------')
# SOCOND APPROCH
print(df.groupby('continent')['human_development_index'].max().sort_values().tail(1))
print("-------------------------------------------------------------------------")
#THIRD APPORCH
print(df.groupby('continent')['human_development_index'].max().sort_values(ascending =False).head(1))
```

```python
# 3: F: FIND WHICH CONTINENT HAS MINIMUM GDP PER CAPITA: 
print(df.groupby('continent')['gdp_per_capita'].min().idxmin())
print('------------------------------------------------------')
# SECOND APPROCH
print(df.groupby('continent')['gdp_per_capita'].min().sort_values().head(1))
print('------------------------------------------------------')
# THIRD APPROCH
print(df.groupby('continent')['gdp_per_capita'].min().sort_values(ascending= False).tail(1))
```

```python
# 4 FILTER THE DATA FRAME: FILTER THE DATAFRAME WHERE CONTINENT IS ASIA AND TOTAL DEATH CASES GREATER THEN 5000

print(df[(df['continent']=='Asia') & (df['total_deaths']>5000)])
```

```python
# 4 FILTER DATAFRAME : FILTERING DATA WHERE GDP PER CAPITA IS GREATER THAN 1500 AND HUMAN DEVELOPMENT INDEX IS MAXIMUM:

print(df[(df['gdp_per_capita']>1500) & (df['human_development_index']==max(df['human_development_index']))])
```

```python
# 4 : FILTERING DATA :
print(df.sort_values(by=['continent','location','date','total_cases','total_deaths','gdp_per_capita','human_development_index']).isin(df))

```

```python
# 4 FILTERING DATASET WHERE TOTAL_CASES EQUAL TO TOTAL DEATHS:

print(df[df['total_cases']==df['total_deaths']])
```

```python
# 5 DATA CLEANINING: (A): ROMOVE ALL DUPLICATES OBSERVATIONS:
df1= df.copy()
df1 = df1.drop_duplicates()
print(df1)
```

```python
# 5: DATA CLEANING: (B): FIND MISSING VALUE IN ALL COLUMNS:
print(f"missing values in all columns\n{df1.isna().sum()}")
print('---------------------------------------------')
# SECOND APPROCH:
print(f"missing values in all columns\n{df1.isnull().sum()}")
```

```python
# 5: DATA CLEANING: (C): REMOVE ALL OBSERVATION WHERE CONTINENT COLUMN VALUE IS MISSING: 
df1 = df1.dropna(how='all',subset=['continent'])
print(f"After Removing missing values\n{df1}")
```

```python
# 5: DATA CLEANING: (D): FILL ALL MISSING VALUES WITH  0:
df1=df1.fillna(0)
print(f"After Filling null values with 0\n{df1}")
```

```python
# 6: DATE TIME FORMATE: (A): CONVERT DATE COLUMN IN DATE TIME FORMATE USING PANDAS. TO DATETIME:
df1['date'] = pd.to_datetime(df1['date'])
print(df1['date'].dtype)
```

```python
# : DATETIME FORMATE: (B): CREATE NEW COLUMN MONTH AFTER EXTRACTING MONTH  DATA FROM DATETIME:
df1['month'] =df1['date'].dt.month
print(df1.head())
```

```python
# 7 DATA AGGREGATION (A): FIND THE MAX VALUE IN ALL COLUMNS USING GROUPBY FUNCTION ON CONTINENT:
print(f"max value in all columns via continent\n{df1.groupby(by='continent').max().reset_index().max()}")
print("--------------------------------------------")
#SECOND APPROCH
print(df1.groupby('continent').max().sort_values(by = 'continent').max())
```

```python
# 7 DATA AGGREGATION (B): STORE THE RESULT IN A NEW DATAFRAME NAMED DF_GROUPBY.(USE DF_GROUPBY DATA FRAME FOR ALL FURTHER ANALYSIS)
df_groupby = df1.groupby(by='continent').max().reset_index()
display(df_groupby)
```

```python
# 8 FEATURE ENGINEERING : (A) CREATE A NEW FEATURE "total_deaths_to_total_case" by "ratio of total_deaths" COLUMNS TOTAL CASES
df_new_data =df.copy()
df_new_data['total_deaths_to_total_case']=(df_new_data['total_deaths']/df_new_data['total_cases'])
df_new_data.head()
```

```python
# 9 DATA VISUALIZATION: (A) PERFORM UNIVARIATE ANALYSIS ON GDP PER CAPITA COLUMN BY PLOTING HISTOGRAM USING SEABORN DIST

plt.figure(figsize=(10,8))
sns.set()
sns.distplot(x=df1['gdp_per_capita'],label="Gdp_Per_Capita",kde = True, color = 'blue',bins = 15)
plt.title("Histogram Plot")
plt.show()
```

```python
from numpy.core.numerictypes import sctype2char
# 9 : DATA VISULIZATION: (B) PLOT A SCATTER PLOT OF TOTAL CASES AND GDP PER CAPITA;

plt.figure(figsize=(10,8))
sns.set()
scatter_plot = plt.scatter(x=df['total_cases'],y=df['gdp_per_capita'],c='r',s=100,edgecolor='b')
plt.xlabel('total_cases')
plt.ylabel('gdp_per_capita')
plt.title('scatter plot')
plt.legend()
plt.show()

```

```python
# 9 DATA VISULIZATION (C): PLOT PAIRPLOT ON DF_GROUPBY DATASET

sns.pairplot(df_groupby, hue='continent',palette='coolwarm')
```

```python
# 9 DATA VISULIZATION (D) BAR_PLOT OF CONTINENT COLUMN WITH TOTAL CASES;

sns.set()
plt.figure(figsize=(10,8))
sns.barplot(x='continent',y='total_cases', data=df)
plt.xlabel('Continent')
plt.ylabel('Total Cases')
plt.title('Total Cases By Continent')
plt.show()
```
